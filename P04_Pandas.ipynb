### Series

A **Series** is a one-dimensional labeled array: values plus an **index** (row labels).

```python
ser = pd.Series([1, 3, 5, 7])
print(type(ser))
ser
```

```python
ser.values   # NumPy array of values
ser.index    # index labels (default: 0, 1, 2, ...)
```

```python
ser2 = pd.Series([1, 3, 5, 7], index=["d", "b", "c", "a"])
ser2
```

```python
# Label-based selection (index labels, not positions)
ser2["a"]
ser2[["a", "b", "d"]]
```

```python
# Boolean filtering
ser2[ser2 > 4]
```

```python
# Vectorized arithmetic (aligned by index)
ser2 * 2
np.exp(ser2)
```

```python
# Membership tests the *index*, not the values
"b" in ser2
```

Build a Series from a Python `dict`. Missing index labels become `NaN`.

```python
sdata = {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000}
ser3 = pd.Series(sdata)
ser3
```

```python
states = ["California", "Ohio", "Oregon", "Texas"]
ser4 = pd.Series(sdata, index=states)
ser4
```

```python
# Missing values: prefer isna / notna (isnull / notnull are aliases)
ser4.isna()
ser4.notna()
```

```python
# Arithmetic aligns on the index; non-overlapping labels become NaN
ser3 + ser4
```

```python
ser4.name = "population"
ser4.index.name = "states"
ser4
```

### DataFrame

A **DataFrame** is a table: rows and columns, each axis has an index. Think of it as a dict of Series sharing the same row index.

```python
data = {
    "state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada"],
    "year": [2000, 2001, 2002, 2001, 2002],
    "pop": [1.5, 1.7, 3.6, 2.4, 2.9],
}
df = pd.DataFrame(data)
print(type(df))
df
```

```python
df = pd.DataFrame(data, columns=["year", "state", "pop"])
df
```

```python
df2 = pd.DataFrame(
    data,
    columns=["year", "state", "pop", "debt"],
    index=["one", "two", "three", "four", "five"],
)
df2
```

```python
# Column access: attribute or brackets (brackets required if name has spaces)
df2["year"]
df2.state
```

### `loc` and `iloc`

- **`loc`**: label-based (row/column names)
- **`iloc`**: integer position-based (0-based)

Prefer `df.loc[row, col]` over chained indexing such as `df[col][row]`.

```python
df2.loc["two"]          # one row (Series)
df2.loc["two", "pop"]     # one scalar
df2.loc["two":"four", ["year", "pop"]]
```

```python
df2.iloc[1]             # second row
df2.iloc[1:3, [0, 2]]     # rows 1–2, columns 0 and 2
```

```python
df2["debt"] = 5.0
df2
```

```python
# Assign a Series aligned on the index (non-matching rows become NaN)
ser = pd.Series([-1.2, -1.5, -1.7], index=["two", "four", "five"])
df2["debt"] = ser
df2
```

```python
df2["eastern"] = df2["state"] == "Ohio"
df2
```

```python
df2.T  # transpose
```

```python
# Index objects are immutable
df2.index
try:
    df2.index[0] = "oneone"
except TypeError as e:
    print(e)
```

### Reindexing

`reindex` returns a **new** object with labels rearranged, added, or removed. Missing entries are `NaN` unless you fill them.

```python
ser1 = pd.Series([4.5, 7.2, -5.3, 3.6], index=["d", "b", "a", "c"])
ser2 = ser1.reindex(["a", "b", "c", "d", "e"])
ser2
```

```python
ser1.reindex(["a", "b", "c", "d", "e"], fill_value=0)
```

```python
# Ordered categories / time series: forward-fill after reindex
ser3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
ser4 = ser3.reindex(range(6)).ffill()
ser3, ser4
```

### `apply` and sorting

```python
rng = np.random.default_rng(0)
frame = pd.DataFrame(
    rng.standard_normal((4, 3)),
    index=["Utah", "Ohio", "Texas", "Oregon"],
    columns=list("abd"),
)
frame
```

```python
# apply: column-wise (axis=0) or row-wise (axis=1)
frame.apply(lambda col: col.max() - col.min(), axis=0)
```

```python
# Element-wise transforms on a DataFrame use .map() (formerly applymap)
frame.map(lambda x: round(x, 2))
```

```python
frame.sort_index()  # sort rows by index
frame.sort_index(axis=1, ascending=False)
```

### Sample price panel (pandas + NumPy)

Build a small **synthetic** daily panel of closing prices and volume—no extra packages or network. Same pandas workflows apply to real CSV/API data later.

```python
rng = np.random.default_rng(42)
dates = pd.bdate_range("2024-01-01", periods=60)
companies = ["AAPL", "IBM", "MSFT", "GOOG"]

# Simulated closing prices (geometric random walk)
close = pd.DataFrame(
    {
        sym: 100 * np.exp(np.cumsum(rng.standard_normal(len(dates)) * 0.012))
        for sym in companies
    },
    index=dates,
)
volume = pd.DataFrame(
    {sym: rng.integers(1_000_000, 8_000_000, size=len(dates)) for sym in companies},
    index=dates,
)
close.head()
```

### Correlation and covariance

```python
volume.head()
```

```python
# Daily percent change (returns); first row is NaN
ptc_close = close.pct_change()
ptc_close.tail()
```

```python
ptc_close["AAPL"].corr(ptc_close["IBM"])
pd.Series.corr(ptc_close["AAPL"], ptc_close["IBM"])
```

```python
ptc_close["AAPL"].cov(ptc_close["IBM"])
ptc_close.corr()
ptc_close.cov()
```

```python
ptc_close.corrwith(ptc_close["IBM"])
ptc_close.corrwith(volume)  # matching column names
```

### Handling missing data

```python
s_data = pd.Series(["aardvark", "artichoke", np.nan, "avocado"])
s_data.isna()
s_data.dropna()
s_data.fillna("N/A")
```

```python
rng = np.random.default_rng(1)
s = pd.Series(rng.standard_normal(21))
s = s.mask(s < 0)  # preferred over s[s < 0] = np.nan for clarity
df = pd.DataFrame(s.to_numpy().reshape(7, 3))
df
```

```python
rng = np.random.default_rng(2)
df = pd.DataFrame(rng.standard_normal((7, 3)))
df = df.where(df > 0)  # keep positives; others become NaN
df
```

```python
df.fillna({0: -0.5, 1: -1})
df.ffill()  # forward-fill (replaces fillna(method='ffill'))
```

```python
df.mean()
df.fillna(df.mean())
```

### Hierarchical (MultiIndex)

```python
rng = np.random.default_rng(3)
data = pd.Series(
    rng.standard_normal(10),
    index=[
        ["a", "a", "a", "b", "b", "b", "c", "c", "d", "d"],
        [1, 2, 3, 1, 2, 3, 1, 2, 2, 3],
    ],
)
data
data.index
```

```python
data["b"]       # all rows whose outer label is 'b'
data["b":"c"]     # slice on the outer level
```

```python
data.loc[:, 2]  # all outer labels, inner level == 2
```

```python
data.unstack()       # Series -> DataFrame (inner index becomes columns)
data.unstack().stack()  # back to Series; NaN where no value existed
```

```python
frame = pd.DataFrame(
    np.arange(12).reshape((4, 3)),
    index=[["a", "a", "b", "b"], [1, 2, 1, 2]],
    columns=[["Ohio", "Ohio", "Colorado"], ["Green", "Red", "Green"]],
)
frame.index.names = ["key1", "key2"]
frame.columns.names = ["state", "color"]
frame
```

```python
frame.swaplevel("key1", "key2")
frame.sum()
```
